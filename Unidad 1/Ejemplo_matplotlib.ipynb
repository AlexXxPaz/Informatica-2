```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0, 3 * np.pi, 200)
y1 = np.sin(x1)
x2 = np.linspace(0, 2 * np.pi, 200)
y2 = np.cos(x2)
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(6,9))
ax1.plot(x1, y1,'v',label="Grafica sen",color="red",linewidth=1)
ax2.plot(x2, y2,label="Grafica cos",color="black",linewidth=3)
ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("Grfica de la funcin sen(x)")
ax1.legend()
ax1.grid()
ax1.text(3.5,0.50,"Esta es una prueba")
ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("Grfica de la funcin cos(x)")
ax2.legend()
ax2.grid()
plt.show()
```
